```python
import matplotlib.pyplot as pit
import numpy as np 
%matplotlib inline
```

```python
fx=lambda x: x**2*np.cos(x)-3*x**3
```

```python
x=np.linspace(-20,20,25)
x2=np.linspace(0,5,5)
y=np.cos(x)
y2=x2**2*np.cos(x2)
pit.plot (x,fx(x))
pit.plot (x2,fx(x2))
```

```python
fx(x)
```

```python
def fx2(x,y):
    z=x*y*np.cos(x/y)-3*x**3
    return z
```

```python
pit.plot(x,fx2(x,y))
```

```python
x3=np.random.rand(10)
fx2(x3,2)
```

```python
def recta(a,m=1,b=0):
    return m*a+b
```

```python
x=np.linspace(-5,5,20)
pit.plot(x,recta(x,3))
pit.plot(x,recta(x,3,-2))
pit.grid()
```

```python
x=np.linspace(-2,2,20)
recta (x)
```
