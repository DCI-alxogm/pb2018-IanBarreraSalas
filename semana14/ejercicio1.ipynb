```python
import matplotlib.pyplot as pit
import numpy as np 
%matplotlib inline
```

```python
def ux(b,h,theta,w):
    z=b+(h*(-(w**2)*theta))
    return z
```

```python
def tx(theta,u,h):
    z2=theta+u*h
    return z2
```

```python
v = np.zeros(1000)
o = np.zeros(1000)
h = .001
vv= 5
v[0]=2
o[0]=10
```

```python
for i in range (1,1000):
    v[i] = ux(v[i-1],h,o[i-1],vv)
    o[i] = tx(o[i-1], v[i-1],h)
```

```python
pit.plot (v)
```

```python
pit.plot(o)
```

```python
pit.plot(v,o)
```
