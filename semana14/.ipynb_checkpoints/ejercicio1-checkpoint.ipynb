```python
import matplotlib.pyplot as pit
import numpy as np 
%matplotlib inline
```

```python
def ux(b,h,theta,w):
    z=b+(h*(-(w**2)*theta))
    return z
```

```python
def tx(theta,u):
    z2=theta+u
    return z2
```

```python
h = np.zeros(1000)
for f in range(0,1000):
    h[f]=f+1
```

```python
v = np.zeros(1000)
o = np.zeros(1000)
vv= 2*3.1416/h
v[0]=200
o[0]=10
```

```python
for i in range (1,1000):
    v[i] = ux(v[i-1],h[i-1],o[i-1],vv[i-1])
    o[i] = tx(o[i-1], v[i-1])
```

```python
print (v)
pit.plot (v)
```

```python
print (o)
pit.plot(o)
```

```python
pit.plot(v,o)
```
